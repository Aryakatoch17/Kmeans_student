```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df= pd.read_csv('/content/student_clustering.csv')
```

```python
print(df.head())
```

## Preprocesing

```python
scaler=StandardScaler()
df_scaled=scaler.fit_transform(df[['cgpa','iq']])
```

```python
kmeans=KMeans(n_clusters=3, random_state=42)
kmeans.fit(df_scaled)
```

```python
df['Cluster']=kmeans.labels_
```

```python
df
```

```python
plt.figure(figsize=(10,7))
plt.scatter(df['cgpa'],df['iq'],c=df['Cluster'], cmap='viridis',s=100)
plt.title('Student Clusters')
plt.xlabel('CGPA')
plt.ylabel('IQ')
```

```python
centers = kmeans.cluster_centers_
centers_unscaled = scaler.inverse_transform(centers)
plt.scatter(centers_unscaled[:, 0], centers_unscaled[:, 1], c='red', s=300, marker='o', edgecolors='black')
plt.show()
```
